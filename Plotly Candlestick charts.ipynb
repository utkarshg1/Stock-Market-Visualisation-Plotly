# Stock Candlestick chart with MA and EMA analysis - Utkarsh Gaikwad

# Getting Stock Data from yfinance

```python
import yfinance as yf
ticker= 'ITC.NS'
df = yf.download(ticker, period='1y',interval='1d')
df.head()
```

# Getting 50 Simple Moving average and 9 Exponential Moving Avergae

```python
df['MA50']=df['Close'].rolling(50).mean()
df['EMA9']=df['Close'].ewm(span=9,adjust=False).mean()
df.tail()
```

```python
import plotly.graph_objects as go

fig1 = go.Figure(data=[go.Candlestick(x=df.index,
                                        open=df['Open'],
                                        high=df['High'],
                                        low=df['Low'],
                                        close=df['Close']),
                        go.Scatter(x=df.index, y=df.MA50, line=dict(color='orange', width=1)),
                        go.Scatter(x=df.index, y=df.EMA9, line=dict(color='blue', width=1))])

title1 = {
         'text': ticker+" Candlestick Chart",
         'y':0.9, # new
         'x':0.5,
         'xanchor': 'center',
         'yanchor': 'top' # new
        }

fig1.update_layout(xaxis_rangeslider_visible=False,
                 title=title1,
                 yaxis_title=ticker+" Stock Price")
fig1.show()
```

```python
df['mean20']=df['Close'].rolling(20).mean()
df['std']=df['Close'].rolling(20).std()
df['Upper'] = df['mean20'] + 2*df['std']
df['Lower'] = df['mean20'] - 2*df['std']
df.tail()
```

```python
fig2 = go.Figure(data=[go.Candlestick(x=df.index,
                                        open=df['Open'],
                                        high=df['High'],
                                        low=df['Low'],
                                        close=df['Close']),
                        go.Scatter(x=df.index, y=df.Upper, line=dict(color='blue', width=1)),
                        go.Scatter(x=df.index, y=df.Lower, line=dict(color='orange', width=1)),
                        go.Scatter(x=df.index, y=df.mean20,line=dict(color='black',width=1))])

title2 = {
         'text': ticker+" Candlestick Chart with Bollinger Band",
         'y':0.9, # new
         'x':0.5,
         'xanchor': 'center',
         'yanchor': 'top' # new
        }

fig2.update_layout(xaxis_rangeslider_visible=False,
                 title=title2,
                 yaxis_title=ticker+" Stock Price")
fig2.show()
```
